## Introduction

Matplotlib is a library that allows the creation of quality figures and graphics through the use of Python.

 - Allows easy and efficient chart creation

 - Allows integration of graphs and figures in a Jupyter Notebook

#### Import Matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
# Show integrated graphics inside jupyter notebook
%matplotlib inline
```

### Graphical representation of data

 - If the plot function is given an array of data, it will use that as the coordinates on the vertical axis, and it will use the index of each data point in the array as the horizontal coordinate.

```python
plt.plot([1,2,5,7,8,3,1])
# plt.show()
plt.savefig("img/gp.png", transparent=False)
```

 - Two matrices can also be provided: one for the horizontal axis, and one for the vertical axis.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

 - The lengths of the axes can be modified so that the figure does not look so tight

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # [xmin, xmax, ymin, ymax]
plt.show()
```

 - The same procedure is followed to paint a mathematical function

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

 - The style of the graph can also be modified to contain more information. 

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

 - Graphics can be overlaid and line styling can be changed

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
# Separating the functions in different lines
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

 - In order to differentiate between both functions, it is always recommended to add a legend

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1")
plt.legend(loc="best") # La situa en la mejor localización
plt.show()
```

 - You can also create two graphs that do not overlap. These graphs are organized in a grid and are called subplots.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g')

plt.show()
```

 - So that the graphs are not so tight, we can make the figure larger.

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize=14)
plt.ylabel("y1", fontsize=14)

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize=14)
plt.ylabel("y2", fontsize=14)

plt.show()
```

### Scatter plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.savefig("img/scatter.png", transparent=False)
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c='red')
plt.scatter(x2, y2, c='blue')
plt.show()
```

### Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211) # 2 rows, 1 columns, 1st subplot
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(212) # 2 rows, 1 columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.95)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.savefig("img/hist.png", transparent=False)
```
